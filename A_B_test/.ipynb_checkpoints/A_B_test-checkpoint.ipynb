```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind
from scipy.stats import chi2_contingency as chi2 

import pandahouse as ph

import seaborn as sns
import matplotlib.pyplot as plt
from pylab import rcParams

from scipy.stats import mannwhitneyu
from scipy.stats import ttest_ind
from scipy.stats import norm
import matplotlib.pyplot as plt

from tqdm.auto import tqdm

plt.style.use('ggplot')
```

```python
groups = pd.read_csv("groups.csv", sep=";") \
        .rename(columns={"id":"user_id", "grp": "group"})

group_add = pd.read_csv("group_add.csv", sep=",") \
        .rename(columns={"id":"user_id", "grp": "group"})

active = pd.read_csv("active_studs.csv", sep=";") \
        .rename(columns={"student_id":"user_id"})

check = pd.read_csv("checks.csv", sep=";") \
        .rename(columns={"student_id":"user_id", "rev": "revenue"})
```

```python
# –û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∏–∑ groups.csv –∏ group_add.csv
all_users = groups.append(group_add) \
    .reset_index(drop=True) \
    .rename(columns={"id":"user_id"})

all_users
```

```python
# –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –≤ active —Å—Ç–æ–ª–±–µ—Ü —Å –ø–æ–º–µ—Ç–∫–æ–π –æ–± –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è 
# (1 - –∑–Ω–∞—á–∏—Ç –∑–∞—Ö–æ–¥–∏–ª –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º—É –≤ –¥–Ω–∏ –ø—Ä–æ–≤–µ–¥–µ–Ω–∏—è —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞)
active["active"] = "active"
active
```

```python
# –û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö –æ –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–Ω–æ—Å—Ç–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –∫ –∫–æ–Ω—Ç—Ä–æ–ª—å–Ω–æ–π –∏–ª–∏ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞–ª—å–Ω–æ–π –≥—Ä—É–ø–ø–µ
# —Å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–µ–π –æ –ø–æ—Å–µ—â–µ–Ω–∏–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã –≤ –¥–Ω–∏ –ø—Ä–æ–≤–µ–¥–µ–Ω–∏—è —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞ (active - –¥–∞, no_active - –Ω–µ—Ç) –∏
# —Å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–µ–π –æ–± –æ–ø–ª–∞—Ç–∞—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ –¥–Ω–∏ –ø—Ä–æ–≤–µ–¥–µ–Ω–∏—è —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞.
df = pd.merge(all_users, active, how="left", on="user_id").fillna("no_active")
df = pd.merge(df, check, how="left", on="user_id")

# –î–æ–±–∞–≤–ª–µ–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞ —Å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–µ–π –æ —Ç–æ–º —Å–æ–≤–µ—Ä—à–∏–ª –ª–∏ –æ–ø–ª–∞—Ç—É –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å
df["payment"] = df["revenue"].apply(lambda x: "pay" if x > 0 else "no_pay")

df
```

```python
# –§–∏–ª—Ç—Ä–∞—Ü–∏—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, –∫–æ—Ç–æ—Ä—ã–µ –ø–æ—Å–µ—â–∞–ª–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º—É –≤ –¥–Ω–∏ –ø—Ä–æ–≤–µ–¥–µ–Ω–∏—è —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞ (active)
active_users = df.query("active == 'active'")[["user_id", "group", "revenue", "payment"]]

active_users
```

```python
# –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ –ø—Ä–æ–≤–µ—Ä–∏–º –±—ã–ª–∏ –ª–∏ –æ–ø–ª–∞—Ç—ã —Å—Ä–µ–¥–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, 
# –∫–æ—Ç–æ—Ä—ã–µ –Ω–µ –∑–∞—Ö–æ–¥–∏–ª–∏ –Ω–∞ —Å–∞–π—Ç –≤ –¥–Ω–∏ –ø—Ä–æ–≤–µ–¥–µ–Ω–∏—è —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞ (active='no_active')

paid_users = df.query("active == 'no_active' & revenue > 0")

paid_users
```

```python
# –û–∫–∞–∑—ã–≤–∞–µ—Ç—Å—è –µ—Å—Ç—å –¥–∞–Ω–Ω—ã–µ –æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è—Ö, –∫–æ—Ç–æ—Ä—ã–µ –Ω–µ –∑–∞—Ö–æ–¥–∏–ª–∏ –Ω–∞ —Å–∞–π—Ç –≤ –¥–Ω–∏ –ø—Ä–æ–≤–µ–¥–µ–Ω–∏—è —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞,
# –Ω–æ –ø—Ä–∏ —ç—Ç–æ–º —Å–æ–≤–µ—Ä—à–∏–ª–∏ –æ–ø–ª–∞—Ç—É. –ù–∞ –º–æ–π –≤–∑–≥–ª—è–¥, —Ç–∞–∫–æ–µ –º–æ–≥–ª–æ –ø—Ä–æ–∏–∑–æ–π—Ç–∏ –ø–æ –¥–≤—É–º –ø—Ä–∏—á–∏–Ω–∞–º:
#     1) –Ω–∞ —Å–∞–π—Ç–µ –µ—Å—Ç—å —Å–∏—Å—Ç–µ–º–∞ –ø–æ–¥–ø–∏—Å–∫–∏ –∏ —É —ç—Ç–∏—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –ø—Ä–æ–∏–∑–æ—à–ª–æ –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –ø—Ä–æ–¥–ª–µ–Ω–∏–µ –ø–æ–¥–ø–∏—Å–∫–∏, 
#     —Ö–æ—Ç—å –Ω–∞ —Å–∞–π—Ç –æ–Ω–∏ –Ω–µ –∑–∞—Ö–æ–¥–∏–ª–∏.
#     2) –ø—Ä–æ–∏–∑–æ—à–ª–∞ –æ—à–∏–±–∫–∞ –≤ —Å–±–æ—Ä–µ –¥–∞–Ω–Ω—ã—Ö –∏ —á–∞—Å—Ç—å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, –∫–æ—Ç–æ—Ä—ã–µ –∑–∞—Ö–æ–¥–∏–ª–∏ –Ω–∞ —Å–∞–π—Ç –≤ –¥–Ω–∏ –ø—Ä–æ–≤–µ–¥–µ–Ω–∏—è —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞, 
#     –±—ã–ª–∏ –ø–æ–º–µ—á–µ–Ω—ã, –∫–∞–∫ no_active. –í —Ç–∞–∫–æ–º —Å–ª—É—á–∞–µ —Å—Ç–æ–∏—Ç —É—á–µ—Å—Ç—å —ç—Ç–∏—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –ø—Ä–∏ –ø—Ä–æ–≤–µ–¥–µ–Ω–∏–∏ A/B —Ç–µ—Å—Ç–∞.
```

```python
# –ü—Ä–æ—Ä–∞–±–æ—Ç–∞–µ–º –≤–∞—Ä–∏–∞–Ω—Ç –Ω–æ–º–µ—Ä 2
```

```python
# –û–±—ä–µ–¥–∏–Ω–∏–º –¥–∞–Ω–Ω—ã–µ –ø–æ –∞–∫—Ç–∏–≤–Ω—ã–º –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º –∏ –Ω–µ –∞–∫—Ç–∏–≤–Ω—ã–º, –Ω–æ —Å–æ–≤–µ—Ä—à–∏–≤—à–∏–º –æ–ø–ª–∞—Ç—É.
active_users = pd.concat([active_users, paid_users[['user_id', 'group', 'revenue', 'payment']]], ignore_index=True)

active_users
```

```python
# –°—Ä–∞–≤–Ω–µ–Ω–∏–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∞–∫—Ç–∏–≤–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π (–∫–æ—Ç–æ—Ä—ã–µ –∑–∞—à–ª–∏ –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º—É –≤ –¥–Ω–∏ –ø—Ä–æ–≤–µ–¥–µ–Ω–∏—è —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞) 
# –≤ —Ç–µ—Å—Ç–æ–≤–æ–π –∏ –∫–æ–Ω—Ç—Ä–æ–ª—å–Ω–æ–π –≥—Ä—É–ø–ø–µ.

sns.histplot(data=active_users[["group", "user_id"]], x="group", shrink=.7)
```

```python
# –°–≤–µ–¥–µ–Ω–∏–µ –∞–∫—Ç–∏–≤–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ –∫—Ä–æ—Å—Å-—Ç–∞–±–ª–∏—Ü—É –ø–æ –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–Ω–æ—Å—Ç–∏ –∫ –∫–æ–Ω—Ç—Ä–æ–ª—å–Ω–æ–π/—Ç–µ—Å—Ç–æ–≤–æ–π –≥—Ä—É–ø–ø–µ 
# –∏ —Ñ–∞–∫—Ç—É –æ–ø–ª–∞—Ç—ã –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ. –ö—Ä–æ—Å—Å-—Ç–∞–±–ª–∏—Ü–∞ –ø–æ–∫–∞–∂–µ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∞–∫—Ç–∏–≤–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å–æ–≤–µ—Ä—à–∏–≤—à–∏—Ö 
# –∏ –Ω–µ —Å–æ–≤–µ—Ä—à–∏–≤—à–∏—Ö –æ–ø–ª–∞—Ç—É –≤ –∫–∞–∂–¥–æ–π –∏–∑ –≥—Ä—É–ø–ø.

payment_group = pd.crosstab(active_users["group"], active_users["payment"])

payment_group
```

```python
# –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–ø–ª–∞—Ç–∏–≤—à–∏—Ö –∏ –Ω–µ –æ–ø–ª–∞—Ç–∏–≤—à–∏—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ —Ç–µ—Å—Ç–æ–≤–æ–π –∏ –∫–æ–Ω—Ç—Ä–æ–ª—å–Ω–æ–π –≥—Ä—É–ø–ø–µ –æ—Ç–ª–∏—á–∞–µ—Ç—Å—è. 
# –û–¥–Ω–∞–∫–æ, –¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥ —Ç–æ–ª—å–∫–æ –ø–æ —Ü–∏—Ñ—Ä–∞–º ‚Äì –Ω–µ –æ—á–µ–Ω—å —Ö–æ—Ä–æ—à–æ. 
# –ù—É–∂–Ω–æ –ø—Ä–∏–º–µ–Ω–∏—Ç—å —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–π —Ç–µ—Å—Ç –∏ –ø–æ–Ω—è—Ç—å —è–≤–ª—è–µ—Ç—Å—è —Ä–∞–∑–Ω–∏—Ü–∞ —Å—Ç–∞—Ç.–∑–Ω–∞—á–∏–º–æ–π.

# –ì–∏–ø–æ—Ç–µ–∑—ã:
# ùêª0 : –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–ø–ª–∞—Ç–∏–≤—à–∏—Ö –∏ –Ω–µ –æ–ø–ª–∞—Ç–∏–≤—à–∏—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ —Ç–µ—Å—Ç–æ–≤–æ–π –∏ –∫–æ–Ω—Ç—Ä–æ–ª—å–Ω–æ–π –≥—Ä—É–ø–ø–µ —Å—Ç–∞—Ç.–∑–Ω–∞—á–∏–º–æ –ù–ï —Ä–∞–∑–ª–∏—á–∞–µ—Ç—Å—è
# ùêª1 : –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–ø–ª–∞—Ç–∏–≤—à–∏—Ö –∏ –Ω–µ –æ–ø–ª–∞—Ç–∏–≤—à–∏—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ —Ç–µ—Å—Ç–æ–≤–æ–π –∏ –∫–æ–Ω—Ç—Ä–æ–ª—å–Ω–æ–π –≥—Ä—É–ø–ø–µ —Å—Ç–∞—Ç.–∑–Ω–∞—á–∏–º–æ —Ä–∞–∑–ª–∏—á–∞–µ—Ç—Å—è
    
# –ü–æ—Å–∫–æ–ª—å–∫—É —Å—Ä–∞–≤–Ω–∏–≤–∞–µ–º –¥–≤–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ (group –∏ payment), –Ω—É–∂–µ–Ω —Ö–∏-–∫–≤–∞–¥—Ä–∞—Ç

stat, p, dof, expected = chi2(payment_group)

stat, p
```

```python
# –ü–æ–ª—É—á–µ–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ p_value > 0.05, —á—Ç–æ –≤—Å–µ —Ä–∞–≤–Ω–æ —Å–ª—É—á–∞–µ –Ω–µ –¥–∞–µ—Ç –æ—Å–Ω–æ–≤–∞–Ω–∏—è –æ—Ç–∫–ª–æ–Ω–∏—Ç—å –ù0,
# —Ç.–µ. –Ω–µ—Ç –æ—Å–Ω–æ–≤–∞–Ω–∏–π —É—Ç–≤–µ—Ä–∂–¥–∞—Ç—å, —á—Ç–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–ø–ª–∞—Ç–∏–≤—à–∏—Ö –∏ –Ω–µ –æ–ø–ª–∞—Ç–∏–≤—à–∏—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π 
# –≤ —Ç–µ—Å—Ç–æ–≤–æ–π –∏ –∫–æ–Ω—Ç—Ä–æ–ª—å–Ω–æ–π –≥—Ä—É–ø–ø–µ —Å—Ç–∞—Ç.–∑–Ω–∞—á–∏–º–æ —Ä–∞–∑–ª–∏—á–∞–µ—Ç—Å—è
```

```python
# –û–¥–Ω–∞–∫–æ, —Å–∞–º–æ –ø–æ —Å–µ–±–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–ø–ª–∞—Ç –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å –Ω–µ –æ—á–µ–Ω—å –∏–Ω—Ñ–æ—Ä–º–∞—Ç–∏–≤–Ω—ã–π. 
# –ü–æ—ç—Ç–æ–º—É –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ —Å—Ä–∞–≤–Ω–∏–º —Å—Ä–µ–¥–Ω–∏–π —á–µ–∫ —Å—Ä–µ–¥–∏ –∞–∫—Ç–∏–≤–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ —Ä–∞–∑—Ä–µ–∑–µ –≥—Ä—É–ø–ø (–∫–æ–Ω—Ç—Ä–æ–ª—å–Ω–æ–π –∏ —Ç–µ—Å—Ç–æ–≤–æ–π).
# –ë—É–¥–µ–º —Å—á–∏—Ç–∞—Ç—å –∞–∫—Ç–∏–≤–Ω—ã–º–∏ —Ç–µ—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π. –∫–æ—Ç–æ—Ä—ã–µ –∑–∞—Ö–æ–¥–∏–ª–∏ –Ω–∞ —Å–∞–π—Ç –≤ –¥–Ω–∏ –ø—Ä–æ–≤–µ–¥–µ–Ω–∏—è —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞ –∏ —Ç–µ—Ö, –∫—Ç–æ –Ω–µ –∑–∞—Ö–æ–¥–∏–ª,
# –Ω–æ —Å–æ–≤–µ—Ä—à–∏–ª –æ–ø–ª–∞—Ç—É.
```

```python
active_users
```

```python
# –†–∞—Å—Å—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, —Å–æ–≤–µ—Ä—à–∏–≤—à–∏—Ö –æ–ø–ª–∞—Ç—É, –≤ —Ä–∞–∑—Ä–µ–∑–µ –≥—Ä—É–ø–ø.
paid_users = active_users \
                .query("payment == 'pay'") \
                .groupby('group', as_index=False) \
                .agg({'user_id': 'count'}) \
                .rename(columns={'user_id': 'paid_users'})

paid_users
```

```python
# –†–∞—Å—Å—á–∏—Ç–∞–µ–º —Å—É–º–º–∞—Ä–Ω—ã–π –¥–æ—Ö–æ–¥ –≤ —Ä–∞–∑—Ä–µ–∑–µ –≥—Ä—É–ø–ø.
total_revenue = active_users \
                    .groupby('group', as_index=False) \
                    .agg({'revenue': 'sum'}) \
                    .rename(columns={'user_id': 'total_revenue'})

total_revenue
```

```python
# –û–±—ä–µ–¥–∏–Ω–∏–º paid_users –∏ total_revenue –∏ —Ä–∞—Å—Å—á–∏—Ç–∞–µ–º —Å—Ä–µ–¥–Ω–∏–π —á–µ–∫ (Average Order Value)
# –¢–∞–∫–∂–µ –ø–æ—Å—Ç—Ä–æ–∏–º boxplot –æ–ø–ª–∞—Ç –≤ —Ä–∞–∑—Ä–µ–∑–µ –≥—Ä—É–ø–ø
data = pd.merge(paid_users, total_revenue, how="left", on="group")

data
```

```python
data['aov'] = data['revenue'] / data['paid_users']

data
```

```python
# –ü–æ—Å—Ç—Ä–æ–∏–º boxplot
revenue_by_groups = active_users.query("payment == 'pay'")[['group', 'revenue']]

sns.boxplot(x="group", y="revenue", data=revenue_by_groups)
```

```python
# –ü–æ —Ç–∞–±–ª–∏—á–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏—è–º –∏ –ø–æ –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏ –≤–∏–¥–Ω–æ, —á—Ç–æ —Å—Ä–µ–¥–Ω–∏–π —á–µ–∫ –¥–ª—è —Ç–µ—Å—Ç–æ–≤–æ–π –∏ –∫–æ–Ω—Ç—Ä–æ–ª—å–Ω–æ–π –≥—Ä—É–ø–ø—ã –æ—Ç–ª–∏—á–∞–µ—Ç—Å—è.
# –í —Ç–µ—Å—Ç–æ–≤–æ–π –≥—Ä—É–ø–ø–µ —Å—Ä–µ–¥–Ω–∏–π —á–µ–∫ –±–æ–ª—å—à–µ
# –ù–µ–æ–±—Ö–æ–¥–∏–º–æ –ø—Ä–æ–≤–µ—Å—Ç–∏ A/B —Ç–µ—Å—Ç –∏ –≤—ã—è—Å–Ω–∏—Ç—å —è–≤–ª—è–µ—Ç—Å—è –ª–∏ —ç—Ç–∞ —Ä–∞–∑–Ω–∏—Ü–∞ —Å—Ç–∞—Ç.–∑–Ω–∞—á–∏–º–æ–π

# –¢–∞–∫–∂–µ –ø–æ –≥—Ä–∞—Ñ–∏–∫—É –≤–∏–¥–∏–º, —á—Ç–æ –µ—Å—Ç—å –≤—ã–±—Ä–æ—Å—ã. –í –¥–∞–Ω–Ω–æ–º —Å–ª—É—á–∞–µ –∏—Ö –º–æ–∂–Ω–æ –æ—Å—Ç–∞–≤–∏—Ç—å, 
# –ø–æ—Å–∫–æ–ª—å–∫—É —ç—Ç–æ –≤–ø–æ–ª–Ω–µ –º–æ–≥–ª–∏ –±—ã—Ç—å —Ä–µ–¥–∫–∏–µ –ø–æ–∫—É–ø–∫–∏ –Ω–∞ —Å—É–º–º—ã –±–æ–ª—å—à–µ —Å—Ä–µ–¥–Ω–µ–≥–æ
```

```python
# –ü–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è revenue –ø–æ –≥—Ä—É–ø–ø–∞–º
A = revenue_by_groups.query("group == 'A'")['revenue']
sns.distplot(A, kde=False, axlabel='Group A')
plt.show()

B = revenue_by_groups.query("group == 'B'")['revenue']
sns.distplot(B, kde=False, axlabel='Group B')
plt.show()
```

```python
# –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –æ—Ç–ª–∏—á–∞—é—Ç—Å—è –æ—Ç –Ω–æ—Ä–º–∞–ª—å–Ω—ã—Ö.
# –ù–æ, –ø–æ—Å–∫–æ–ª—å–∫—É –∑–Ω–∞—á–µ–Ω–∏–π –≤ –æ–±–µ–∏—Ö –≤—ã–±–æ—Ä–∫–∞—Ö (A –∏ B) –º–Ω–æ–≥–æ (–±–æ–ª–µ–µ 30) –º–æ–∂–Ω–æ –ø—Ä–æ–≤–µ—Å—Ç–∏ A/B –±–µ–∑ –ø—Ä–∏–≤–µ–¥–µ–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö –∫ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–º—É –≤–∏–¥—É.
# –ü—Ä–∏–º–µ–Ω–∏–º t-test.
```

```python
ttest_ind(A, B)
```

```python
# t-test –ø–æ–∫–∞–∑–∞–ª p-value < 0.05 (pvalue=0.0025070167572044375), 
# —á—Ç–æ –¥–∞–µ—Ç –æ—Å–Ω–æ–≤–∞–Ω–∏—è –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ä–∞–≤–µ–Ω—Å—Ç–≤–µ —Å—Ä–µ–¥–Ω–∏—Ö. 
# –¢.–µ. –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥, —á—Ç–æ —Å—Ä–µ–¥–Ω–∏–µ —á–µ–∫–∏ –¥–≤—É—Ö –≤—ã–±–æ—Ä–æ–∫ (–¥–≤—É—Ö –≥—Ä—É–ø–ø –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π) —Å—Ç–∞—Ç.–∑–Ω–∞—á–∏–º–æ —Ä–∞–∑–ª–∏—á–∞—é—Ç—Å—è. 
```

```python
# –î–ª—è —É–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç–∏ –Ω–æ—Ä–º–∞–ª–∏–∑—É–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –≤ –æ–±–µ–∏—Ö –≤—ã–±–æ—Ä–∫–∞—Ö (A –∏ B), –ø—Ä–∏–º–µ–Ω–∏–≤ –ª–æ–≥–∞—Ä–∏—Ñ–º–∏—Ä–æ–≤–∞–Ω–∏–µ –∏ –ø–æ–≤—Ç–æ—Ä–∏–º A/B —Ç–µ—Å—Ç.
```

```python
A_log = np.log(A)
B_log = np.log(B)
```

```python
# –ü–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è revenue –ø–æ –≥—Ä—É–ø–ø–∞–º
sns.distplot(A_log, kde=False, axlabel='Group A_log')
plt.show()

sns.distplot(B_log, kde=False, axlabel='Group B_log')
plt.show()
```

```python
ttest_ind(A_log, B_log)
```

```python
# t-test —Å–Ω–æ–≤–∞ –ø–æ–∫–∞–∑–∞–ª p-value < 0.05 (pvalue=0.0004482400492255275), 
# —á—Ç–æ –¥–∞–µ—Ç –æ—Å–Ω–æ–≤–∞–Ω–∏—è –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ä–∞–≤–µ–Ω—Å—Ç–≤–µ —Å—Ä–µ–¥–Ω–∏—Ö. 
# –¢.–µ. –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥, —á—Ç–æ —Å—Ä–µ–¥–Ω–∏–µ —á–µ–∫–∏ –¥–≤—É—Ö –≤—ã–±–æ—Ä–æ–∫ (–¥–≤—É—Ö –≥—Ä—É–ø–ø –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π) —Å—Ç–∞—Ç.–∑–Ω–∞—á–∏–º–æ —Ä–∞–∑–ª–∏—á–∞—é—Ç—Å—è. 
```

### –û—Ç–≤–µ—Ç—ã –Ω–∞ –≤–æ–ø—Ä–æ—Å—ã:

**1. –ù–∞ –∫–∞–∫–∏–µ –º–µ—Ç—Ä–∏–∫–∏ –í—ã —Å–º–æ—Ç—Ä–∏—Ç–µ –≤ —Ö–æ–¥–µ –∞–Ω–∞–ª–∏–∑–∞ –∏ –ø–æ—á–µ–º—É?**

–ü–æ—Å–∫–æ–ª—å–∫—É —Ä–µ—á—å –∏–¥–µ—Ç –æ —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–∏ –Ω–æ–≤–æ–≥–æ –º–µ—Ö–∞–Ω–∏–∑–º–∞ –æ–ø–ª–∞—Ç—ã, —Ç–æ –ø—Ä–∏ –ø—Ä–æ–≤–µ–¥–µ–Ω–∏–∏ A/B —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è –Ω—É–∂–Ω–æ –æ–±—Ä–∞—Ç–∏—Ç—å –≤–Ω–∏–º–∞–Ω–∏–µ –Ω–∞ —Ñ–∞–∫—Ç –æ–ø–ª–∞—Ç—ã –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–º —É—Å–ª—É–≥ –Ω–∞ —Å–∞–π—Ç–µ –∏ —Å—Ä–µ–¥–Ω–∏–π —á–µ–∫. –≠—Ç–æ –∏ –µ—Å—Ç—å –æ—Å–≤–Ω–æ–Ω—ã–µ –º–µ—Ç—Ä–∏–∫–∏ –¥–ª—è —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è –Ω–æ–≤–æ–≥–æ –º–µ—Ö–∞–Ω–∏–∑–º–∞ –æ–ø–ª–∞—Ç—ã.

**!–í –æ–±—â–µ–º –Ω–∞ –æ—Å–Ω–æ–≤–∞–Ω–∏–∏ –∏–º–µ—é—â–∏—Ö—Å—è –¥–∞–Ω–Ω—ã—Ö –º–æ–∂–Ω–æ –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ —Ä–∞—Å—Å—á–∏—Ç–∞—Ç—å —Å–ª–µ–¥—É—é—â–∏–µ –º–µ—Ç—Ä–∏–∫–∏:**
- **–î–æ–ª—è –∞–∫—Ç–∏–≤–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π (Active Users Rate)** ‚Äì –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∑–∞—Ö–æ–¥–∏–≤—à–∏—Ö –Ω–∞ —Å–∞–π—Ç –≤ –¥–Ω–∏ –ø—Ä–æ–≤–µ–¥–µ–Ω–∏—è —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞ –¥–µ–ª–µ–Ω–Ω–æ–µ –Ω–∞ –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π.
- **–û–±—â–∏–π –¥–æ—Ö–æ–¥ (Summary Revenue), —Å—Ä–µ–¥–Ω–∏–π —á–µ–∫ (Average Order Value)** –≤ –¥–Ω–∏ –ø—Ä–æ–≤–µ–¥–µ–Ω–∏—è —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞.
- **–ö–æ–Ω–≤–µ—Ä—Å–∏—è –∞–∫—Ç–∏–≤–Ω–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –≤ –æ–ø–ª–∞—Ç—É (CR Active)** ‚Äì –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∞–∫—Ç–∏–≤–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å–æ–≤–µ—Ä—à–∏–≤—à–∏—Ö –æ–ø–ª–∞—Ç—É –¥–µ–ª–µ–Ω–Ω–æ–µ –Ω–∞ –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∞–∫—Ç–∏–≤–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π.
- **–ö–æ–Ω–≤–µ—Ä—Å–∏—è –Ω–µ –∞–∫—Ç–∏–≤–Ω–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –≤ –æ–ø–ª–∞—Ç—É (CR No Active)** ‚Äì –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–µ –∞–∫—Ç–∏–≤–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å–æ–≤–µ—Ä—à–∏–≤—à–∏—Ö –æ–ø–ª–∞—Ç—É –¥–µ–ª–µ–Ω–Ω–æ–µ –Ω–∞ –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–µ –∞–∫—Ç–∏–≤–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π.
- **ARPU** ‚Äì –æ–±—â–∏–π –¥–æ—Ö–æ–¥ –∑–∞ –≤—Ä–µ–º—è –ø—Ä–æ–≤–µ–¥–µ–Ω–∏—è —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞ –¥–µ–ª–µ–Ω–Ω—ã–π –Ω–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∞–∫—Ç–∏–≤–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π.
- **ARPAU** ‚Äì —Å—Ä–µ–¥–Ω–∏–π –ø–ª–∞—Ç–µ–∂ –¥–µ–ª–µ–Ω–Ω—ã–π –Ω–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∞–∫—Ç–∏–≤–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π.

**2. –ò–º–µ—é—Ç—Å—è –ª–∏ —Ä–∞–∑–ª–∏—á–∏—è –≤ –ø–æ–∫–∞–∑–∞—Ç–µ–ª—è—Ö –∏ —Å —á–µ–º –æ–Ω–∏ –º–æ–≥—É—Ç –±—ã—Ç—å —Å–≤—è–∑–∞–Ω—ã?**

–ï—Å–ª–∏ —Å–º–æ—Ç—Ä–µ—Ç—å –Ω–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–ø–ª–∞—Ç —Å—Ä–µ–¥–∏ –∞–∫—Ç–∏–≤–Ω—ã—Ö –≤ –¥–Ω–∏ –ø—Ä–æ–≤–µ–¥–µ–Ω–∏—è —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, —Ç–æ –∫–∞–∂–µ—Ç—Å—è, —á—Ç–æ –µ—Å—Ç—å —Ä–∞–∑–ª–∏—á–∏—è. –ù–æ –¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥—ã —Ç–æ–ª—å–∫–æ –Ω–∞ —Ä–∞–∑–Ω–∏—Ü–µ –≤ —á–∏—Å–ª–∞—Ö - —ç—Ç–æ –Ω–µ –≤–µ—Ä–Ω—ã–π –ø–æ–¥—Ö–æ–¥. –ù—É–∂–Ω–æ –ø—Ä–æ–≤–µ—Å—Ç–∏ A/B —Ç–µ—Å—Ç –∏ –ø–æ–Ω—è—Ç—å, —è–≤–ª—è–µ—Ç—Å—è –ª–∏ —Ä–∞–∑–Ω–∏—Ü–∞ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º–æ–π.

**3. –Ø–≤–ª—è—é—Ç—Å—è –ª–∏ —ç—Ç–∏ —Ä–∞–∑–ª–∏—á–∏—è —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º—ã–º–∏?**

A/B —Ç–µ—Å—Ç (—Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–π —Ç–µ—Å—Ç –°—Ç—å—é–¥–µ–Ω—Ç–∞(ttest_ind)) —Ä–∞–∑–Ω–∏—Ü—ã —Å—Ä–µ–¥–Ω–∏—Ö —á–µ–∫–æ–≤ –≤ —Ä–∞–∑—Ä–µ–∑–µ –≥—Ä—É–ø–ø (–∫–æ–Ω—Ç—Ä–æ–ª—å–Ω–æ–π –∏ —Ç–µ—Å—Ç–æ–≤–æ–π) –ø–æ–∫–∞–∑–∞–ª p-value < 0.05, —á—Ç–æ –¥–∞–µ—Ç –æ—Å–Ω–æ–≤–∞–Ω–∏—è —Å–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥, —á—Ç–æ —Å—Ä–µ–¥–Ω–∏–µ —á–µ–∫–∏ –¥–≤—É—Ö –≤—ã–±–æ—Ä–æ–∫ (–¥–≤—É—Ö –≥—Ä—É–ø–ø –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π) —Å—Ç–∞—Ç.–∑–Ω–∞—á–∏–º–æ —Ä–∞–∑–ª–∏—á–∞—é—Ç—Å—è. –¢.–µ. —Å—Ä–µ–¥–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Ç–µ—Å—Ç–æ–≤–æ–π –≥—Ä—É–ø–ø—ã —Å—Ä–µ–¥–Ω–∏–π —á–µ–∫ —Å—Ç–∞—Ç.–∑–Ω–∞—á–∏–º–æ –≤—ã—à–µ.

**4. –°—Ç–æ–∏—Ç –ª–∏ –∑–∞–ø—É—Å–∫–∞—Ç—å –Ω–æ–≤—É—é –º–µ—Ö–∞–Ω–∏–∫—É –Ω–∞ –≤—Å–µ—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π?**

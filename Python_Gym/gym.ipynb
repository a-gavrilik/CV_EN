***Search for emissions by Age column***

```python
# Calculate 0.25 and 0.75 quantiles
q1 = df.Age.quantile(0.25)
q3 = df.Age.quantile(0.75)

# Calculate the interquartile range
iqr = q3 - q1

# Calculate the lower and upper limit of the interquartile range
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)

# Remove outliers from df and write the result without outliers in remove_outliers
remove_outliers = df[df.Age.between(lower_bound, upper_bound, inclusive=True)]

# Combine df and remove_outliers, remove duplicates, and get the outliers themselves
pd.concat([df, remove_outliers]).drop_duplicates(keep=False).sort_values('Age')
```

**Conclusion**:
* in this case the values of ages of people who in our reality (if the data on fitness centers in Russia or the CIS countries) do not yet attend a fitness center (fairly young people 18-20 years old) or think that a fitness center is not for them anymore (adults 38 - 41) got into the emissions
* another reason is that the marketing of the fitness center is not directed towards these categories of customers (perhaps because these categories of people are less solvent or less motivated/interested), so they are few and they are in the outliers
* in any case these outliers may be considered acceptable, people 18-20 and 38-41 also visit the fitness center, just not many of them

* there are no emissions, but there is a value of 13 (13 months until the end of the subscription). This is strange, because the maximum subscription validity is 12 months. 
* let's additionally check the data for entries where the expiration date (Month_to_end_contract) exceeds the validity period of the subscription (Contract_period).

```python
df[df['Month_to_end_contract'] > df['Contract_period']]
```

**Conclusion**:
* There are data on clients (there are many of them), whose term to the end of the current subscription in months exceeds by 1 (if it exceeds, then always by 1) the period of validity of the subscription. For example, the validity period is 12 months. - The period to the end of 13 months. Or the subscription validity period is 6 months. - The period to the end of 7 months. 
* This is probably the case in the automatic calculation of the subscription expiration date, if the subscription itself is issued at the junction of calendar months. For example, on March 31, 2020 a yearly subscription is issued for the period from April 1, 2020 to April 1, 2021. If the system counts the subscription expiration date from the day of registration (instead of the start day) to the end day in full months (i.e. from March 2020 to April 2021), then the subscription expiration date may be 1 month longer.
* In my opinion, this is a mistake. You should correct the data by condition: if the term to expiration date is 1 month longer than the validity period, then the term to expiry date should be reduced by 1
    


```python
# Correct the data by the condition: 
# if the period to the end is 1 month longer than the validity period, then reduce the period to the end by 1

df.loc[(df['Month_to_end_contract'] > df['Contract_period']), 'Month_to_end_contract'] = df['Month_to_end_contract'] - 1
```

***section of customer churn data. And answer the questions:***

- Is there a meaningful difference in average age for clients who have dropped out and clients who have retained?
- Is there a meaningful difference in the average amount spent on other services for churned and retained clients?
- Is there a meaningful difference in the average number of months until the end of a subscription for dropping out and retained customers?
- Is there a significant difference in the average Lifetime for clients who have dropped out and clients who have stayed in?
- Is there a meaningful difference in average total attendance for dropping off and retaining customers?
- Is there a meaningful difference in average attendance over the past month for dropped and retained customers?
    
***To do this, formulate null and alternative hypotheses for each question, for example:***

- H0: the average age of clients for which the churned (Churn=1) and not churned (Churn=0) are the same 
- H1: the average age of clients for whom the fact of churn is fixed (Churn=1) and for whom the fact of churn is not fixed (Churn=0) does not coincide 

***Since the cost of error in this test is not high, let's set the statistical significance threshold alpha = 0.05.***.

***We will perform a two-sample two-sided test since we have two independent samples.***

```python
# fit-transform immediately learns and converts the data
X_scaled = scaler.fit_transform(cl[['Age', 
                                    'Avg_additional_charges_total',  
                                    'Contract_period', 
                                    'Month_to_end_contract', 
                                    'Lifetime', 
                                    'Avg_class_frequency_total', 
                                    'Avg_class_frequency_current_month']])
```

```python
X_scaled
```

```python
from sklearn.cluster import KMeans
```

```python
# Construct an elbow graph for X_scaled

# Create a list for inertia
k_inertia = []
# Set the cluster range
ks = range(1, 20)

for k in ks:
    clf_kmeans = KMeans(n_clusters=k)
    clusters_kmeans = clf_kmeans.fit_predict(X_scaled)
    # Add the inertia of each model to the list
    k_inertia.append(clf_kmeans.inertia_)
    
plt.plot(ks, k_inertia)
plt.plot(ks, k_inertia ,'ro')
```

```python
# Explicitly specify the number of clusters based on the elbow graph
kmeans = KMeans(n_clusters=3)

# fit_predict is trained on the data and assigns a cluster to each object
clusters = kmeans.fit_predict(X_scaled)

# Let's add a column with the cluster number for each client to cl_churn_1
cl['Cluster'] = clusters

cl
```

```python
# Calculate the mean values of the features in each cluster
cl[['Age', 
    'Avg_additional_charges_total',  
    'Contract_period', 
    'Month_to_end_contract', 
    'Lifetime', 
    'Avg_class_frequency_total', 
    'Avg_class_frequency_current_month',
    'Cluster']].groupby('Cluster').mean()
```

**Conclusion:**
* Cluster=0 - clients from this cluster are distinguished by the fact that they are slightly older on average. They also have the highest Lifetime and the highest average attendance.
* Cluster=1 - clients from this cluster are distinguished by the fact that they, on average, have the minimum age, purchase subscriptions for the minimum period and use other services of the fitness center the least. They also have the minimum Lifetime and average attendance.
* Cluster=2 - clients from this cluster are distinguished by the fact that they purchase subscriptions for a long period and on average spend the most on other services of the fitness center. They also have a fairly high Lifetime (almost the same as the clients with Cluster=0), but the average attendance is lower than the clients with Cluster=0.
